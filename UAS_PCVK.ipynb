```python
import cv2 as cv
import numpy as np
import os
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Read Image
directory = '/content/drive/MyDrive/PCVK/Tubes Kel.5'
image_files = [file for file in os.listdir(directory) if file.endswith((".jpg", ".jpeg", ".JPG"))]

# Pretrained feature
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/PCVK/facedet/haarcascade_frontalface_default.xml')
```

```python
# Loop melalui setiap file gambar
for image_file in image_files:
    # Baca gambar
    image_path = os.path.join(directory, image_file)
    sim = cv2.imread(image_path)

    # Deteksi wajah
    sim1 = sim.copy()
    roi_wajah = cascade_wajah.detectMultiScale(sim)

    for (x, y, w, h) in roi_wajah:
        cv2.rectangle(sim1, (x, y), (x+w, y+h), (0, 0, 255), 15)

    # Tampilkan gambar asli dan gambar dengan deteksi wajah
    plt.figure(figsize=(10, 10))
    plt.subplot(121), plt.imshow(cv2.cvtColor(sim, cv2.COLOR_BGR2RGB))
    plt.title('Original Image'), plt.axis('off')
    plt.subplot(122), plt.imshow(cv2.cvtColor(sim1, cv2.COLOR_BGR2RGB))
    plt.title('Image with Face Detection'), plt.axis('off')
    plt.show()
```
