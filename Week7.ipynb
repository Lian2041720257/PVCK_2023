```python
# Import Library
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image as im
```

```python

# Load Image
img = cv.imread('/content/drive/MyDrive/PCVK/female (1).png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

Konvolusi fungsi Library

```python
def convolution2d(image, kernel, stride, padding):

    image_padded = np.pad(image, padding, mode='constant')

    padded_width, padded_height = image_padded.shape
    kernel_width, kernel_height = kernel.shape

    output_width = (padded_width - kernel_width) // stride + 1
    output_height = (padded_height - kernel_height) // stride + 1

    img_convolved = np.zeros((output_width, output_height))

    for i in range(output_width):
        for j in range(output_height):
            img_convolved[i, j] = np.sum(image_padded[i * stride:i * stride + kernel_width, j * stride:j * stride + kernel_height] * kernel)

    cv2_imshow(img_convolved)
```

Sharpen

```python
# Image Sharpen
kernel_sharpen = np.array([[0, -1, 0],
                          [-1, 5, -1],
                          [0, -1, 0]])

convolution2d(img_gray, kernel_sharpen, 1, 2)
```

Emboss

```python
# Image Emboss
kernel_emboss = np.array([[-2, -1, 0],
                          [-1, 1, 1],
                          [0, 1, 2]])

convolution2d(img_gray, kernel_emboss, 1, 2)
```

Left Sobel Edge Detection

```python
# Image Sobel Edge Detection
kernel_sobel = np.array([[1,0,-1],
                         [2,0,-2],
                         [1,0,-1]])

convolution2d(img_gray, kernel_sobel, 1, 2)
```

Canny Edge Detection

```python
# Image Canny Edge Detection
kernel_canny = np.array([[-1,-1,-1],
                         [-1,8,-1],
                         [-1,-1,-1]])
convolution2d(img_gray, kernel_canny, 1, 2)
```

Prewitt Edge Detection

```python
# Image Prewitt Edge Detection
kernel_prewitt = np.array([[-1,-1,-1],
                         [0,0,0],
                         [1,1,1]])
convolution2d(img_gray, kernel_prewitt, 1, 2)
```
